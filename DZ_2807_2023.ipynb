```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kc_house_data.csv',encoding = 'cp1251')
```

```python
df['price'].describe()
```

```python
plt.hist(df['sqft_living'], bins=50)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Количество домов')
plt.show()
```

```python
plt.hist(df['yr_built'], bins=50)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.show()
```

По графику видно, что большинство домов были построены в период с 1950 по 2010 годы.

```python
plt.scatter(df['sqft_living'], df['price'])
plt.title('Зависимость цены от площади дома')
plt.xlabel('Площадь дома')
plt.ylabel('Цена')
plt.show()
```

Чем больше площадь, тем выше цена.

```python
waterfront_prices = df.groupby('waterfront')['price'].mean()
plt.bar(waterfront_prices.index, waterfront_prices.values)
plt.title('Зависимость цены от наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Средняя цена')
plt.xticks([0,1], ['Нет', 'Да'])
plt.show()
```

Дома с видом на набережную стоят значительно дороже, чем дома без него.
